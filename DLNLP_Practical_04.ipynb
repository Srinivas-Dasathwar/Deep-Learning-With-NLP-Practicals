```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

df = pd.read_csv('/content/superheroes_nlp_dataset.csv')
df['name'].fillna('', inplace=True)

document = df['name'].tolist()
print(document)

Counter_Vectorizer = CountVectorizer()
count_matrix = Counter_Vectorizer.fit_transform(document)
```

```python
Counter_Vectorizer = CountVectorizer()
count_matrix = Counter_Vectorizer.fit_transform(document)
X = Counter_Vectorizer.fit_transform(document)
```

```python
word_count_array = X.toarray()
```

```python
feature_names = Counter_Vectorizer.get_feature_names_out()
```

```python
print("Feature Names (Vocabulary):",feature_names)
print("Word Count Vectors:\n",word_count_array)
```
