```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize,sent_tokenize
Filepath = '/content/Unemployment_Rate_M_Total.txt'
def read_dataset(Filepath):
  with open(Filepath, 'r') as file:
    data = file.read()
    return data
```

```python
text = read_dataset('/content/Unemployment_Rate_M_Total.txt')
sentences=sent_tokenize(text)
sentences
```

```python
for i, sentence in enumerate(sentences):
  print(f"sentence No {i}: {sentence}")
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
for word in text.split():
  print(ps.stem(word))
```

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
for word in text.split():
  print(lemmatizer.lemmatize(word,pos="v"))
  print(lemmatizer.lemmatize(word,pos="n"))
  print(lemmatizer.lemmatize(word,pos="a"))
  print(lemmatizer.lemmatize(word,pos="r"))
```
